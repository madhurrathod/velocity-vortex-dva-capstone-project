# 02 Cleaning

Notebook goals:
- standardize columns
- treat `unknown` as missing for selected fields
- convert `pdays = 999` into a meaningful business indicator
- create segmentation columns for analysis and Tableau

```python
from pathlib import Path
import numpy as np
import pandas as pd

BASE_DIR = Path.cwd().resolve().parents[0]
RAW_FILE = BASE_DIR / 'data' / 'raw' / 'bank_marketing' / 'bank-additional' / 'bank-additional' / 'bank-additional-full.csv'
df = pd.read_csv(RAW_FILE, sep=';')
df.columns = [col.strip().lower().replace('.', '_').replace('-', '_') for col in df.columns]
for col in ['job', 'marital', 'education', 'default', 'housing', 'loan']:
    df[col] = df[col].replace('unknown', np.nan)
df['subscribed_flag'] = (df['y'] == 'yes').astype(int)
df['contacted_before'] = np.where(df['pdays'] == 999, 'No', 'Yes')
df['pdays_clean'] = df['pdays'].replace(999, np.nan)
df['age_group'] = pd.cut(df['age'], bins=[0, 24, 34, 44, 54, 64, 120], labels=['18-24', '25-34', '35-44', '45-54', '55-64', '65+'])
df.head()
```
