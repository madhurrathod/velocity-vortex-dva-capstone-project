# 04 Statistical Analysis

Notebook goals:
- test whether subscription is associated with contact type
- test whether previous campaign outcome is associated with subscription
- compare call duration between subscribers and non-subscribers

```python
from pathlib import Path
import pandas as pd
from scipy.stats import chi2_contingency, ttest_ind

BASE_DIR = Path.cwd().resolve().parents[0]
df = pd.read_csv(BASE_DIR / 'data' / 'processed' / 'bank_marketing_cleaned.csv')

contact_table = pd.crosstab(df['contact'], df['subscribed'])
chi2_contingency(contact_table)
```

```python
poutcome_table = pd.crosstab(df['poutcome'], df['subscribed'])
chi2_contingency(poutcome_table)
```

```python
yes_duration = df.loc[df['subscribed_flag'] == 1, 'duration']
no_duration = df.loc[df['subscribed_flag'] == 0, 'duration']
ttest_ind(yes_duration, no_duration, equal_var=False)
```
