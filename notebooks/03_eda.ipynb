# 03 EDA

Notebook goals:
- measure overall subscription rate
- compare conversion across contact types, months, jobs, and age groups
- identify campaign fatigue patterns

```python
from pathlib import Path
import pandas as pd

BASE_DIR = Path.cwd().resolve().parents[0]
df = pd.read_csv(BASE_DIR / 'data' / 'processed' / 'bank_marketing_cleaned.csv')
round(df['subscribed_flag'].mean() * 100, 2)
```

```python
(df.groupby('contact')['subscribed_flag'].mean() * 100).round(2).sort_values(ascending=False)
```

```python
(df.groupby('age_group')['subscribed_flag'].mean() * 100).round(2).sort_values(ascending=False)
```

```python
campaign_view = (df.groupby('campaign_band')['subscribed_flag'].mean() * 100).round(2)
campaign_view
```
